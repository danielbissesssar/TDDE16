```python
#imports
!pip install --upgrade pip
!pip install transformers
!pip install datasets
!pip install sentencepiece
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, AutoModelForSeq2SeqLM, T5ForConditionalGeneration, AutoModelWithLMHead
from datasets import load_dataset
```

```python
tokenizerG = AutoTokenizer.from_pretrained("valhalla/t5-base-squad")

modelG = AutoModelWithLMHead.from_pretrained("valhalla/t5-base-squad")
```

```python
#load data
dataset = load_dataset("squad")
dataset = dataset['validation']
dataset = dataset[0:1000]
goldAnswers = []
for i in range(len(dataset['answers'])):
  goldAnswers.append(dataset['answers'][i]['text'])
```

```python
def get_answerG(question, context):
  input_text = "question: %s  context: %s </s>" % (question, context)
  features = tokenizerG([input_text], return_tensors='pt')

  out = modelG.generate(input_ids=features['input_ids'], 
               attention_mask=features['attention_mask'])
  
  return tokenizerG.decode(out[0])
```

```python
predictedAnswersG = []
for i in range(len(dataset['question'])):
  print(i)
  predictedAnswersG.append(get_answerG(dataset['question'][i],dataset['context'][i]))
```

```python
tokenizerGbase = AutoTokenizer.from_pretrained("t5-base")
modelGbase = AutoModelWithLMHead.from_pretrained("t5-base")
def get_answerGbase(question, context):
  input_text = "question: %s  context: %s </s>" % (question, context)
  features = tokenizerGbase([input_text], return_tensors='pt')

  out = modelGbase.generate(input_ids=features['input_ids'], 
               attention_mask=features['attention_mask'])
  
  return tokenizerGbase.decode(out[0])
```

```python
predictedAnswersGbase = []
for i in range(len(dataset['question'])):
  print(i)
  predictedAnswersGbase.append(get_answerGbase(dataset['question'][i],dataset['context'][i]))
```

```python
from transformers import pipeline
pipelineE = pipeline('question-answering', "thatdramebaazguy/roberta-base-squad")
```

```python
predictedAnswersE = []
for i in range(len(dataset['question'])):
  print(i)
  predictedAnswersE.append(pipelineE(question=dataset['question'][i], context=dataset['context'][i])['answer'])
```

```python
def compute_exact_match(prediction, truth):
    return int(prediction == truth)

def compute_f1(prediction, truth):
    pred_tokens = prediction.split()
    truth_tokens = truth.split()
    
    if len(pred_tokens) == 0 or len(truth_tokens) == 0:
        return int(pred_tokens == truth_tokens)
    
    common_tokens = set(pred_tokens) & set(truth_tokens)
    
    if len(common_tokens) == 0:
        return 0
    
    prec = len(common_tokens) / len(pred_tokens)
    rec = len(common_tokens) / len(truth_tokens)
    
    return 2 * (prec * rec) / (prec + rec)
```

```python
from difflib import SequenceMatcher
f1E = 0
f1G = 0
f1Gbase = 0
EME = 0
EMG = 0
EMGbase = 0
SME = 0
SMG = 0
SMGbase = 0
f1comp = 0
EMcomp = 0
SMcomp = 0
f1compBase = 0
EMcompBase = 0
SMcompBase = 0
for i in range(len(goldAnswers)):
  f1E = f1E + max((compute_f1(predictedAnswersE[i],answer)) for answer in goldAnswers[i])
  f1G = f1G + max((compute_f1(predictedAnswersG[i],answer)) for answer in goldAnswers[i])
  f1Gbase = f1Gbase + max((compute_f1(predictedAnswersGbase[i],answer)) for answer in goldAnswers[i])
  EME = EME + max((compute_exact_match(predictedAnswersE[i],answer)) for answer in goldAnswers[i])
  EMG = EMG + max((compute_exact_match(predictedAnswersG[i],answer)) for answer in goldAnswers[i])
  EMGbase = EMGbase + max((compute_exact_match(predictedAnswersGbase[i],answer)) for answer in goldAnswers[i])
  SME = SME + max((SequenceMatcher(None, predictedAnswersE[i], answer).ratio()) for answer in goldAnswers[i])
  SMG = SMG + max((SequenceMatcher(None, predictedAnswersG[i], answer).ratio()) for answer in goldAnswers[i])
  SMGbase = SMGbase + max((SequenceMatcher(None, predictedAnswersGbase[i], answer).ratio()) for answer in goldAnswers[i])
  f1comp = f1comp + compute_f1(predictedAnswersE[i],predictedAnswersG[i])
  EMcomp = EMcomp + compute_exact_match(predictedAnswersE[i],predictedAnswersG[i])
  SMcomp = SMcomp + SequenceMatcher(None, predictedAnswersE[i], predictedAnswersG[i]).ratio()
  f1compBase = f1compBase + compute_f1(predictedAnswersE[i],predictedAnswersGbase[i])
  EMcompBase = EMcompBase + compute_exact_match(predictedAnswersE[i],predictedAnswersGbase[i])
  SMcompBase = SMcompBase + SequenceMatcher(None, predictedAnswersE[i], predictedAnswersGbase[i]).ratio()

f1E = f1E / len(goldAnswers)
f1G = f1G / len(goldAnswers)
f1Gbase = f1Gbase / len(goldAnswers)
EME = EME / len(goldAnswers)
EMG = EMG / len(goldAnswers)
EMGbase = EMGbase / len(goldAnswers)
SME = SME / len(goldAnswers)
SMG = SMG / len(goldAnswers)
SMGbase = SMGbase / len(goldAnswers)
f1comp = f1comp / len(goldAnswers)
EMcomp = EMcomp / len(goldAnswers)
SMcomp = SMcomp / len(goldAnswers)
f1compBase = f1compBase / len(goldAnswers)
EMcompBase = EMcompBase / len(goldAnswers)
SMcompBase = SMcompBase / len(goldAnswers)
print(f1E, EME, SME)
print(f1G, EMG, SMG)
print(f1Gbase, EMGbase, SMGbase)
print(f1comp, EMcomp, SMcomp)
print(f1compBase, EMcompBase, SMcompBase)

```
